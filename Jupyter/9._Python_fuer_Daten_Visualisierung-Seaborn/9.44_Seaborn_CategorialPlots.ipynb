```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.barplot("sex", "total_bill", data=tips, estimator=np.std)
```

```python
sns.countplot("sex", data=tips)
```

```python
sns.boxplot(data=tips, palette="rainbow", orient="h")
```

```python
sns.boxplot("day", "total_bill", hue="smoker", data=tips, palette="coolwarm")
```

```python
sns.violinplot("day", "total_bill", data=tips, hue="sex", split=True, palette="Set1")
```

```python
sns.stripplot("day", "total_bill", data=tips, jitter=True, hue="sex", palette="Set1", dodge=True)
```

```python
sns.swarmplot("day", "total_bill", data=tips, hue="sex", palette="Set1", dodge=True)
```

```python
sns.violinplot("tip", "day", data=tips, palette="rainbow")
sns.swarmplot("tip", "day", data=tips, color="black", size=3)
```

```python
sns.factorplot("sex", "total_bill", data=tips, kind="bar")
```

```python
sns.barplot("sex", "total_bill", data=tips, estimator=np.mean)
```
