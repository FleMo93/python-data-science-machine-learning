```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, "r")
plt.xlabel("X Achse")
plt.ylabel("Y Achse")
plt.title("Diagramm Titel")
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, "r--")
plt.subplot(1,2,2)
plt.plot(y, x, "g*-")
```

```python
af = plt.figure()
axes = af.add_axes([0.1, 0.1, 0.5, 0.5])
axes.plot(x, y, "b")
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Überschrift")
```

```python
af = plt.figure()
axes1 = af.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = af.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y, "b")
axes1.set_xlabel("X")
axes1.set_ylabel("Y")
axes1.set_title("GROSSE")

axes2.plot(y, x, "r")
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("KLEIN")
```

```python
diag, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_title("Titel")
    ax.plot(x, y)

plt.tight_layout()
```

```python
axes
```

```python
diag = plt.figure(figsize=(8,4), dpi=100)

ax = diag.add_axes([0,0,1,1])
ax.plot(x, y)
```
